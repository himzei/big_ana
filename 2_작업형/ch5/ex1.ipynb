```python
import pandas as pd 

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

train.shape, test.shape
```

```python
target = train.pop("Credit_Score")

train.shape, test.shape
```

```python
data = pd.concat([train, test])
data.shape
```

```python
cols = train.select_dtypes(include="object").columns
train[cols].describe()
```

```python
data_oh = pd.get_dummies(data)
data_oh.shape

train = data_oh.iloc[:len(train)].copy()
test = data_oh.iloc[len(train):].copy()

train.shape
train.head(3)
```

```python
# 검증 데이터 나누기 

from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train, 
    target,
    test_size=0.2,
    random_state=2025 
)

X_train.shape
X_val.shape
```

```python
# 머신러닝 학습 및 평가 

from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb

# 랜덤포레스트 
rf = RandomForestClassifier(random_state=2025)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_val)


# light GBM
lgbmc = lgb.LGBMClassifier(random_state=2025, verbose=-1)
lgbmc.fit(X_train, y_train)
pred_lgb = lgbmc.predict(X_val)
```

```python
# 평가
from sklearn.metrics import f1_score

# 랜덤포레스트 평가
result_rf = f1_score(y_val, pred_rf, average="macro")
print(result_rf)

# lightGBM 평가
result_lgb = f1_score(y_val, pred_lgb, average="macro")
print(result_lgb)
```

```python
# 예측 및 결과 파일 생성
pred = rf.predict(test)

submit = pd.DataFrame({
    "pred": pred
})
submit.to_csv("result.csv", index=False)
```
